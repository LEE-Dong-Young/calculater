```python
ab_a = []
for a in ab :
    ab_a.append(a.text)
```

```python
df
```

```python
soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child(1) > td.asc') 
# avg가 안되는 이유... 말씀해주셨는데, 이해못함
```

```python
soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr:nth-child(1) > td') 
# 뒤에를 지워서 이렇게 확인. 4번째에 있져? 왜 4번째지.. ㅋㅋㅋㅋ
```

```python
avg = soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody > tr > td:nth-child(4)') 
```

```python
avg_a = []
for a in avg :
    avg_a.append(a.text)
```

```python
df['AVG'] = avg_a
```

```python
df
```

```python
df['타석'] = df['타석'].astype(int)
```

```python
df['안타'] = df['안타'].astype(int)
```

```python
df['AVG'] = df['AVG'].astype(float)
```

```python
df['AVG2'] = df['안타'] / df['타석']
```

```python
df['AVG2'] = round(df['AVG2'],3)
```

```python
df
```

```python
df['팀명'].value_counts()
```

```python
df = df.rename({'타석' : 'AB', '안타' : 'H'}, axis=1)
```

```python
df
```

```python
df.to_csv('Hitter2022.csv', index=False, encoding='euc-kr')
```
